### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])
bar_heights = [5, 7, 3, 8, 4]
scatter_x = np.random.rand(10) * 10
scatter_y = np.random.rand(10) * 10
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 25, 20]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Chart
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label="Line Plot")
axs[0, 0].set_title('Line Chart')
axs[0, 0].legend()

# Bar Chart
axs[0, 1].bar(x, bar_heights, color='r', label="Bar Chart")
axs[0, 1].set_title('Bar Chart')
axs[0, 1].legend()

# Scatter Plot
axs[1, 0].scatter(scatter_x, scatter_y, color='g', marker='x', label="Scatter Plot")
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].legend()

# Pie Chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['cyan', 'magenta', 'yellow', 'gray'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Show plots
plt.show()

```
